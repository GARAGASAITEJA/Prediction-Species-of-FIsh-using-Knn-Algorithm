```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression

path = "C:\\Users\\VAIO\\Desktop\\PYQT5\\ml1\\LogisticRegression\\Fish Weight estimation\\Fish.csv"

data = pd.read_csv(path)
print(data)

inputs = data.drop('Species','columns')
output = data.drop(['Weight','Length1','Length2','Length3','Height','Width'],'columns')
print(inputs)
print(output)
###
x_train,x_test,y_train,y_test=train_test_split(inputs,output,train_size=0.8)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
model=LogisticRegression()#Intialising the model
model.fit(inputs,output)
y_predic=model.predict(x_test)
print(y_predic)#predicted value from trained model
print(y_test)# actual data from dataset
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predic)
print(cm)
```
